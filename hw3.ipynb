## Статистические эвристики

##Реализуйте фильтрацию аномалий на основе расчитанных статистик [Q1 - (1.5 * IQR), Q3 + (1.5 * IQR)], [Q1 - (3 * IQR), Q3 + (3 * IQR)]

```python
# Функция для расчета
def stats_method(df, variable):
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    inner_fence = 1.5 * iqr
    outer_fence = 3 * iqr

    inner_fence_le = q1 - inner_fence
    inner_fence_ue = q3 + inner_fence

    outer_fence_le = q1 - outer_fence
    outer_fence_ue = q3 + outer_fence

    outliers = []
    for index, x in enumerate(df[variable]):
        if x <= inner_fence_le or x >= inner_fence_ue:
            outliers.append(1)
        else:
            outliers.append(0)
    return outliers
```

```python
outliers = stats_method(df, "Temperature")
```

```python
# Добавляем новый столбец с аномальными точками
df_static=df.copy()
df_static['outlier'] = outliers
```

## Кластеризация

##Используйте Feature-Engineering для создания признаков (лаги, скользящие статистики, признаки даты/времени, etc)


```python
# Feature Engineering
df_fe = df.copy()

df_fe['second'] = df_fe['datetime'].dt.second
df_fe['minute'] = df_fe['datetime'].dt.minute
df_fe['hour'] = df_fe['datetime'].dt.hour
df_fe['day'] = df_fe['datetime'].dt.day

# Скользящие статистики
df_fe['Temperature_mean'] = df_fe['Temperature'].rolling(window=3).mean()
df_fe['Temperature_std'] = df_fe['Temperature'].rolling(window=3).std()
df_fe['Voltage_mean'] = df_fe['Voltage'].rolling(window=3).mean()
df_fe['Voltage_std'] = df_fe['Voltage'].rolling(window=3).std()

# Lags
df_fe['Lag_Pressure'] = df_fe['Pressure'].shift(1)
df_fe['Lag_Current'] = df_fe['Current'].shift(1)
df_fe['Lag_Thermocouple'] = df_fe['Thermocouple'].shift(1)

# Удаление NaN значений после скользящих статистик
df_fe.dropna(inplace=True)
```

##Реализуйте фильтрацию аномалий на основе модели кластеризации (можно взять любой алгоритм и библиотеку, которые вам понравились)


```python
# Выбор признаков для кластеризации
features = df_fe[['hour','minute','Temperature', 'Voltage', 'Temperature_mean', 'Temperature_std', 'Voltage_mean', 'Voltage_std', 'Lag_Pressure','Lag_Current','Lag_Thermocouple']]

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Кластеризация
kmeans = KMeans(n_clusters=2, random_state=42)
df_fe['cluster'] = kmeans.fit_predict(features_scaled)

# Предположим, что аномалии будут в меньшем кластере
cluster_counts = df_fe['cluster'].value_counts()
anomaly_cluster = cluster_counts.idxmin()

# Помечаем аномалии
df_fe['anomaly_kmeans'] = np.where(df_fe['cluster'] == anomaly_cluster, 1, 0)
```

##Визуализируйте результаты работы модели, оцените качество визуально


```python
hv.Points(
    df_fe.query("anomaly_kmeans == 0")).opts(
        size=8,
        color='#f80'
    ) * hv.Curve(df_fe).opts(
        opts.Curve(
            title="Визуалиция ряда с аномальными точками",
            xlabel="Время",
            ylabel="",
            height=500,
            responsive=True,
            show_grid=True,
        )
    )
```

##Оцените качество, построив confusion matrix для предсказаний и исходных данных

```python
if 'anomaly' in df_fe.columns:
    cm = confusion_matrix(df_fe['anomaly'], df_fe['anomaly_kmeans'])
    print("Матрица путаницы:")
    print(cm)

    # Визуализация матрицы путаницы
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Anomaly'],
                yticklabels=['Normal', 'Anomaly'])
    plt.ylabel('True label')
    plt.xlabel('Предсказанная метка')
    plt.title('Матрица путаницы')
    plt.show()
```

##Описание матрицы ошибок:

- True Negative (TN): 561 — модель правильно классифицировала 561 нормальную точку как нормальную.
- False Positive (FP): 126 — модель ошибочно классифицировала 126 нормальных точек как аномалии.
- False Negative (FN): 14 — модель не обнаружила 14 аномальных точек, классифицировав их как нормальные.
- True Positive (TP): 391 — модель правильно классифицировала 391 аномальную точку как аномалию.

##Интерпретация результатов:

- True Positive (TP) и True Negative (TN): Модель корректно обнаружила 391 аномальную точку и 561 нормальную точку, что свидетельствует о хорошей способности модели выделять аномалии и распознавать нормальные точки.
- False Positive (FP): Количество ложноположительных классификаций (126) указывает на склонность модели к ошибочной классификации нормальных точек как аномалий, что может привести к избыточным срабатываниям.
- False Negative (FN): Модель пропустила всего 14 аномалий, что указывает на высокую чувствительность модели к обнаружению аномалий, что важно в задачах мониторинга и предсказания сбоев.

```python
# Отчет о качестве классификации для модели, основанной на метках истинных значений и предсказанных аномалий
print(classification_report(df_fe['anomaly'], df_fe['anomaly_kmeans']))
```

## Isolation Forest

## Реализуйте фильтрацию аномалий на основе модели Isolation Forest

```python
# Isolation Forest
df_if = df.copy()

features = df_if[['Accelerometer1RMS', 'Accelerometer2RMS', 'Current', 'Pressure', 'Temperature', 'Thermocouple', 'Voltage', 'Volume Flow RateRMS']]
model = IsolationForest(
    random_state=42,
    contamination=0.25,
    n_estimators=200,
    max_samples=0.7
)

df_if['anomaly_if'] = model.fit_predict(features)
```

```python
# Преобразуем -1 в 1 (аномалия) и 1 в 0 (норма)
df_if['anomaly_if'] = np.where(df_if['anomaly_if'] == -1, 1, 0)
```

## Визуализируйте результаты работы модели, оцените качество визуально

```python
hv.Points(
    df_if.query("anomaly_if == 1")).opts(
        size=8,
        color='#f80'
    ) * hv.Curve(df_if).opts(
        opts.Curve(
            title="Визуалиция ряда с аномальными точками",
            xlabel="Время",
            ylabel="",
            height=500,
            responsive=True,
            show_grid=True,
        )
    )
```

## Оцените качество, построив confusion matrix для предсказаний и исходных данных

```python
if 'anomaly' in df_if.columns:
    cm = confusion_matrix(df_if['anomaly'], df_if['anomaly_if'])
    print("Матрица путаницы:")
    print(cm)

    # Визуализация матрицы путаницы
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Anomaly'],
                yticklabels=['Normal', 'Anomaly'])
    plt.ylabel('True label')
    plt.xlabel('Предсказанная метка')
    plt.title('Матрица путаницы')
    plt.show()
```

##Описание матрицы ошибок:

- True Negative (TN): 530 — модель правильно классифицировала 530 нормальных точек как нормальные.
- False Positive (FP): 159 — модель ошибочно классифицировала 159 нормальных точек как аномалии.
- False Negative (FN): 290 — модель не обнаружила 290 аномальных точек, классифицировав их как нормальные.
- True Positive (TP): 115 — модель правильно классифицировала 115 аномальных точек как аномалии.

##Интерпретация результатов:

- True Positive (TP) и True Negative (TN): Модель корректно определила 530 нормальных точек и 115 аномальных точек, что указывает на некоторую способность модели обнаруживать аномалии, но с ограниченной точностью.
- False Positive (FP): Модель ошибочно пометила 159 нормальных точек как аномалии, что может указывать на избыточную чувствительность модели к нормальным точкам.
- False Negative (FN): Модель пропустила 290 аномальных точек, что говорит о недостаточной чувствительности модели к выявлению всех аномалий.

```python
# Отчет о качестве классификации для модели, основанной на метках истинных значений и предсказанных аномалий
print(classification_report(df_if['anomaly'], df_if['anomaly_if']))
```

## Соберите агрегированные результаты по всем экспериментам в отдельный датафрейм и сравните между собой, сделайте выводы

```python
# Получаем отчеты классификации для каждого метода
report_if = classification_report(df_if['anomaly'], df_if['anomaly_if'], output_dict=True)
report_static = classification_report(df_static['anomaly'], df_static['outlier'], output_dict=True)
report_kmeans = classification_report(df_fe['anomaly'], df_fe['anomaly_kmeans'], output_dict=True)

# Преобразуем каждый отчет в DataFrame и добавляем название метода
df_report_if = pd.DataFrame(report_if).transpose().assign(Method="Isolation Forest")
df_report_static = pd.DataFrame(report_static).transpose().assign(Method="Static Method")
df_report_kmeans = pd.DataFrame(report_kmeans).transpose().assign(Method="KMeans")

# Объединяем отчеты в один DataFrame
df_comparison = pd.concat([df_report_if, df_report_static, df_report_kmeans], axis=0)

# Упорядочиваем столбцы для удобства
df_comparison = df_comparison.reset_index().rename(columns={'index': 'Metric'})
df_comparison = df_comparison[['Method', 'Metric', 'precision', 'recall', 'f1-score', 'support']]

# Выводим результирующий DataFrame для сравнения
print(df_comparison)
```

## Выводы:

- Лучший метод: **KMeans** — он показал наилучший баланс между точностью и полнотой для обоих классов, что делает его предпочтительным для задач, требующих как точного распознавания нормальных точек, так и обнаружения аномалий.

- **Isolation Forest** подходит для задач, где важно правильно классифицировать нормальные точки, но требует доработки для улучшения чувствительности к аномалиям.

- **Static Method** полностью игнорирует аномалии и лучше всего подходит для задач, где аномалии не критичны.
